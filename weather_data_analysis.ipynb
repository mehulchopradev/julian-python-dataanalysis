```python
from glob import glob
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
```

```python
path = '/Users/mehul.chopra/Documents/personal/training/data-analysis-data/weather_data_input/weather_data_set_*'
files = glob(path)
print(files)
```

```python
dfs = [pd.read_csv(file, delimiter='|', header=None, 
            names=['station_no', 'year', 'datemonth', 'hourminute', 'temperature']) for file in files]
data = pd.concat(dfs, ignore_index=True)
data
```

Max temperature on yearly basis

```python
d1 = data[['year', 'temperature']]
d1
```

```python
d2 = d1.ffill()
d2
```

```python
d3 = d2.groupby('year')
d3
```

```python
d4 = d3.max()
d4
```

```python
d4.plot(kind='line')
plt.show()
```

```python
d5 = data[['station_no','year','temperature']]
d5
```
