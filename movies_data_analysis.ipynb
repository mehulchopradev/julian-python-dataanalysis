```python
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
```

```python
users_path = '/Users/mehul.chopra/Documents/personal/training/data-analysis-data/movielens/users.dat'
users = pd.read_csv(users_path, delimiter='::', names=['user_id','gender','age','occ','zipcode'], header=None)
users
```

```python
movies_path = '/Users/mehul.chopra/Documents/personal/training/data-analysis-data/movielens/movies.dat'
movies = pd.read_csv(movies_path, header=None, delimiter='::', names=['movie_id','title','genres'])
movies
```

```python
ratings_path = '/Users/mehul.chopra/Documents/personal/training/data-analysis-data/movielens/ratings.dat'
ratings = pd.read_csv(ratings_path, header=None, delimiter='::', names=['user_id','movie_id','ratings','timestamp'])
ratings
```

```python
data = pd.merge(pd.merge(ratings, movies), users)
data
```

Popular films (more than 500 people have rated for it)

```python
d1 = data['title']
d1
```

```python
d2 = d1.value_counts()
d2
```

```python
d3 = d2[d2 > 500]
popular_films = d3.index
popular_films
```

```python
d4 = data[['title', 'ratings', 'gender']]
d4
```

```python
d5 = d4.pivot_table('ratings', aggfunc='mean', index='title', columns=['gender'])
d5
```

```python
d6 = d5.loc[popular_films]
d6
```

```python
d7 = d6.sort_values(by='M', ascending=False)
d8 = d7.head(10)
d8.plot(kind='barh')
plt.show()
```

```python
d7['diff'] = d7['M'] - d7['F']
d7['diff'] = d7['diff'].abs()
d10 = d7.sort_values(by='diff', ascending=False)
d10.head(10)['diff'].plot(kind='barh')
plt.show()
```
